## Step 2: Launch the Demo

Start a local web server to serve the demo application.


````python
import http.server
import socketserver
import threading
import webbrowser
from pathlib import Path
import time

# Configuration
PORT = 8080
src_dir = project_root / "src"

# Change to src directory
os.chdir(src_dir)

class MyHTTPRequestHandler(http.server.SimpleHTTPRequestHandler):
    def log_message(self, format, *args):
        # Suppress default logging
        pass

def start_server():
    with socketserver.TCPServer(("", PORT), MyHTTPRequestHandler) as httpd:
        print(f"\nğŸš€ Server started on port {PORT}")
        print(f"\nğŸ“‚ Serving from: {src_dir}")
        print(f"\nğŸŒ Access the demo at: http://localhost:{PORT}")
        print("\nâš ï¸  To stop the server, interrupt the kernel (Kernel > Interrupt)\n")
        httpd.serve_forever()

# Start server in background thread
server_thread = threading.Thread(target=start_server, daemon=True)
server_thread.start()

# Wait a moment for server to start
time.sleep(1)

# Open browser automatically
try:
    webbrowser.open(f'http://localhost:{PORT}')
    print("âœ… Browser opened automatically")
except:
    print("âš ï¸  Could not open browser automatically. Please open the URL manually.")

print("\nğŸ’¡ The demo is now running! Interact with it in your browser.")
print("\nğŸ“– Features to explore:")
print("   - Chat with customizable system prompts")
print("   - Function calling demonstrations")
print("   - Roleplay with different personas")
print("   - Reasoning challenge problems")
````
